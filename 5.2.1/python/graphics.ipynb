```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import interp1d
import seaborn as sns
from scipy.stats import linregress
from scipy.optimize import curve_fit
import sympy
from heapq import nlargest

import PythonGraphMod

# import math
```

```python
fd1 = pd.read_excel("../data/loc_V_4.ods")

U1 = np.array(fd1['V'])        # напряжение В
I1 = np.array(fd1['I']) * 1000 # ток мкА

fig1, ax1 = PythonGraphMod.CreateSimpleGraph()

U_dence1, I_dence1 = PythonGraphMod.InterpolateSet(U1, I1, 'cubic', U1[0], U1[-1])

ax1.scatter(U1, I1, color='black', marker='o', s=8, alpha=1)
ax1.plot(U_dence1, I_dence1, color='blue', alpha=0.75, lw=1.5, ls='-')

## Пики графика
# I[8] = 274 мкА
# I[26] = 419 мкА
# dU = 17.59 = 17.6 В

# рисуем вертикальную линию, отмечаем точку пересечения
ax1.vlines( U1[8], I1.min(), I1[8], ls='-.', color='red', alpha=0.75)
ax1.vlines( U1[26], I1.min(), I1[26], ls='-.', color='red', alpha=0.75)
ax1.scatter(U1[8], I1[8], color='black', marker='o', s=15, alpha=1)
ax1.scatter(U1[26], I1[26], color='black', marker='o', s=15, alpha=1)

# сетка
ax1.minorticks_on()
ax1.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
ax1.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

# оформление
plt.title('Вольтамперная характиристика')
plt.ylabel('I, мкА')
plt.xlabel('U, В')
ax1.text(5, 350, r'$\Delta V \approx 17.6$ В', 
    size = 8,
    bbox=dict(facecolor='white', edgecolor='black', pad=10.0))

fig1.savefig("../images/loc_V_4.pdf")
```
