```python
import math

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from scipy.interpolate import interp1d
import seaborn as sns

from scipy.stats import linregress

import PythonGraphMod
```

```python
# import data file
df_Data_exp_T = pd.read_excel("../data/Te(Ip).ods")

# import data from data file (.ods)
Ip = np.array(df_Data_exp_T['Ip'])
Te = np.array(df_Data_exp_T['Te'])

# calculate a params of tangents
res = linregress(Ip, Te)

# create figure and graph
fig, ax = PythonGraphMod.CreateSimpleGraph()

# create diagrams
ax.scatter(Ip, Te, c = 'black', marker = 'o')
ax.plot(Ip, res.intercept + res.slope * Ip, color='green', alpha=.75, lw=1.5, ls='--')

# diagram desing
# plt.title ('')
plt.xlabel(r'$I_p$, мА')
plt.ylabel(r'$T_e$, эВ')
PythonGraphMod.AddGreed(ax)

plt.show()
fig.savefig('../images/Te(Ip).png')
```

```python
# import data file
df_Data_exp_n = pd.read_excel("../data/ne(Ip).ods")

# import data from data file (.ods)
Ip = np.array(df_Data_exp_n['Ip'])
ne = np.array(df_Data_exp_n['ne'])

# calculate a params of tangents
res = linregress(Ip, ne)

# create figure and graph
fig1, ax1 = PythonGraphMod.CreateSimpleGraph()

# create diagrams
ax1.scatter(Ip, ne, c = 'black', marker = 'o')
ax1.plot(Ip, res.intercept + res.slope * Ip, color='green', alpha=.75, lw=1.5, ls='--')

# diagram desing
# plt.title ('')
plt.xlabel(r'$I_p$, мА')
plt.ylabel(r'$n_e \cdot 10^{10}, cm^{-3}$')
PythonGraphMod.AddGreed(ax1)

plt.show()
fig1.savefig('../images/ne(Ip).png')
```
