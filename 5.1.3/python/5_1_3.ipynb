```python
fd1 = pd.read_excel("./data/data1.ods")

Uc1 = np.array(fd1['Uc'])         # ускоряющая разность потенциалов
Ia1 = np.array(fd1['Ui']) / 10000 # ток анода (напряжение анода, деленное на 10 кОм)

fig1, ax1 = PythonGraphMod.CreateSimpleGraph()

U_dence1, I_dence1 = PythonGraphMod.InterpolateSet(Uc1, Ia1, 'cubic', Uc1[0], Uc1[-1])

ax1.scatter(Uc1, Ia1, color='black', marker='o', s=8, alpha=1)
ax1.plot(U_dence1, I_dence1, color='blue', alpha=0.75, lw=1.5, ls='-')

# сетка
ax1.minorticks_on()
ax1.grid(which = 'major', color = 'black', linewidth = 1, linestyle = '-', alpha = 0.75)
ax1.grid(which = 'minor', color = 'grey', linewidth = 1, linestyle = '--', alpha = 0.5)

# # оформление
plt.title('Вольтамперная характиристика')
plt.ylabel('I, А')
plt.xlabel('U, В')
# ax1.text(5, 350, r'$\Delta V \approx 17.6 \pm 0.02$ В', 
#     size = 8,
#     bbox=dict(facecolor='white', edgecolor='black', pad=10.0))

# fig1.savefig("../images/graph1.pdf")
```
