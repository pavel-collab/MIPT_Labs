```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
V_R = np.array([3.52, 5.35, 7.14, 8.90, 10.53]) # мВ
V_a = np.array([0.44, 0.65, 0.85, 1.07, 1.255])

fig = plt.figure

lin = linregress(V_R, V_a)

plt.scatter(V_R, V_a)
plt.plot(V_R, lin.intercept + lin.slope*V_R)
plt.grid()
plt.ylabel(r'$V_{ср}$ (на катушке) мВ')
plt.xlabel(r'$V_R$ (на сопративлении) мВ')

print('intercept b = (%0.2lf +- %0.2lf) мВ' %(lin.intercept, lin.intercept_stderr))
print('slope k = (%0.3lf +- %0.3lf)' %(lin.slope, lin.stderr))

plt.savefig('../images/calibration.pdf')
```
